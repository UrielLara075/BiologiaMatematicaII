## Análisis del umbral crítico

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv('Fire_v3 UmbralSeparacion-table.csv', skiprows=6)

table_max_per_try = table.groupby(['[run number]', 'Distancia'])['(burned-trees / initial-trees)'].max().reset_index()

table_mean_per_distance = table_max_per_try.groupby('Distancia')['(burned-trees / initial-trees)'].mean().reset_index()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(table_max_per_try['Distancia'], table_max_per_try['(burned-trees / initial-trees)'], alpha=0.3, label="Simulaciones individuales")
plt.plot(table_mean_per_distance['Distancia'], table_mean_per_distance['(burned-trees / initial-trees)'], marker='o', linestyle='-', color='red', label="Promedio por distancia")
plt.xlabel('Distancia (%)')
plt.ylabel('Proporción de bosque quemado')
plt.title('Detección del umbral crítico de distancia')
plt.legend()
plt.grid(True)

plt.show()
```
